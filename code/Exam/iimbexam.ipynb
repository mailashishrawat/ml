```python
import numpy as np
import matplotlib.pyplot as plt

# Generating Xi values
Xi_values =np.arange(0, 10)   # You can adjust the range as per your requirement
#Xi_values =np.arange(100, 801, 100)
B0=-1.796 
B1=-.096
# Calculating corresponding Yi values
z=B0 + B1 * Xi_values

Yi_values = np.exp(z)/(1+np.exp(z))

print(np.exp(B1))
# Print Yi values
for i in range(len(Xi_values)):
    print(f"For Xi = {Xi_values[i]}, Yi = {Yi_values[i]}")
    
# Plotting the logarithmic chart
plt.figure(figsize=(8, 6))
plt.plot(Xi_values, Yi_values, label='Yi = np.exp(B0 + B1 * Xi_values)/(1+np.exp(B0 +B1 * Xi_values))', color='blue')
plt.title('Logarithmic Chart between Yi and -1.796 - 0.096Xi')
plt.xlabel('Xi')
plt.ylabel('Yi')
plt.ylim(0, 1) 
#plt.yscale('log')  # Setting y-axis to logarithmic scale
plt.grid(True)
plt.legend()
plt.show()
```

```python
def calculate_receptive_field(num_layers, kernel_size, stride):
    """
    Calculate the global receptive field for consecutive convolution layers.

    Args:
        num_layers (int): Number of convolution layers.
        kernel_size (int): Kernel size (assumed to be the same for all layers).
        stride (int): Stride (assumed to be the same for all layers).

    Returns:
        int: Global receptive field.
    """
    RF = 1  # Receptive field at input layer
    for _ in range(num_layers):
        RF += (kernel_size - 1) * stride
    return RF

# Example usage
num_layers = 4
kernel_size = 3
stride = 1
receptive_field = calculate_receptive_field(num_layers, kernel_size, stride)
print("Global Receptive Field:", receptive_field)  # Output: 11
```
