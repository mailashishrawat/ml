```python
#compound poisson distribution is a probability distribution that results from the sum of a random number of independent and identically distributed random variables,
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
rate_per_hour = 8  # Poisson rate (lambda)
mean_withdrawal = 50  # Mean of withdrawal amount (per event)
variance_withdrawal = 400  # Variance of withdrawal amount (per event)
num_hours = 5  # Number of hours to simulate

# Simulating number of withdrawals per hour using Poisson distribution
withdrawals_per_hour = np.random.poisson(rate_per_hour, num_hours)

# Simulating withdrawal amounts for each withdrawal event
# Standard deviation is sqrt of variance
std_withdrawal = np.sqrt(variance_withdrawal)

# For each hour, simulate withdrawal amounts and sum them up
total_withdrawals_per_hour = [
    np.sum(np.random.normal(mean_withdrawal, std_withdrawal, num_withdrawals))
    for num_withdrawals in withdrawals_per_hour
]

# Plotting the results
plt.figure(figsize=(10, 6))
sns.lineplot(x=np.arange(num_hours), y=total_withdrawals_per_hour, marker="o", color="blue")

# Annotate each point with the corresponding value
for i, total_withdrawal in enumerate(total_withdrawals_per_hour):
    plt.text(i, total_withdrawal, f'{total_withdrawal:.0f}', ha='center', va='bottom', fontsize=8)

plt.title('Compound Poisson Distribution: Withdrawals Over Time', fontsize=16)
plt.xlabel('Hour', fontsize=12)
plt.ylabel('Total Withdrawals', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#poisson distribution 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson

# Parameters
lambda_param = 12  # Change this to your lambda value
x = np.arange(0, 20)  # Possible outcomes (range of x values)
y = poisson.pmf(x, lambda_param)  # Poisson Probability Mass Function (PMF)

# Plotting
plt.figure(figsize=(8, 6))
sns.lineplot(x=x, y=y, marker="o", color="blue")
plt.title(f'Poisson Distribution (Î» = {lambda_param})', fontsize=16)
plt.xlabel('Number of events', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.grid(True)
plt.show()
```

Markov decision process 
A Markov decision process is a controlled stochastic process of representing and solving problems where there is uncertainty and sequential decision determines the result. To describe a typical MDP problem (Sigaud and Buffet, 2013), the following five-element tuple is used (S, A, T, p, r):
S â€” the state space;
A â€” the set of all possible actions;
T â€” the set of time steps where decisions need to be made;
p() â€” the state transition probability function;
r() â€” the reward function defined on state transitions.
For each time step t in T, the agent is in a current state St. An action At chosen in this current state affects the transition to the next time step t+1 and next state St+1.
Reward rt is obtained by the agent for this transition from St to St+1. The reward rt can be received at the immediate time t or accumulated over the period t and t+1. The same is illustrated in Fig. 2. This process continues iteratively until â€˜end conditionâ€™ is met.

Almost all reinforcement learning algorithms are based on estimating value functions--functions of states (or of state-action pairs) that estimate how good it is for the agent to be in a given state (or how good it is to perform a given action in a given state). The notion of "how good" here is defined in terms of future rewards that can be expected, or, to be precise, in terms of expected return. Of course the rewards the agent can expect to receive in the future depend on what actions it will take. Accordingly, value functions are defined with respect to particular policies.

The objective of the policy iteration algorithm is to, Find an optimal policy that will maximize the Value function
