```python
# Classification using Decision tree: 

# Decision Tree Learning or Classification Trees are a collection of divide and conquer problem-solving strategies that use a tree-like structure to predict the value of an outcome variable. The original data is divided into subsets in this process. This is done to create more homogenous groups at the children nodes. It is one of the most powerful predictive analytics techniques used for generating business rules.

# Classification and Regression Tree (CART) is one of the classification tree techniques. CART is an umbrella term; we will call it a classification tree if the outcome variable value is discrete and a regression tree if the outcome variable value is continuous.

# Classification tree uses various impurity measures such as the Gini Impurity Index and Entropy to split the nodes. Regression tree, on the other hand, splits the node that minimizes the Sum of Squared Errors (SSE).
import pandas as pd 
from sklearn import metrics
import statsmodels.api as sm
import numpy as np
from sklearn.model_selection import train_test_split


df=pd.read_csv("../../inputdata/German_Credit_Data.csv")
#df.info()

xfeature=list(df.columns)
xfeature.remove("status")
# print("printing xfeatures")
# print(xfeature)
creditsdf= pd.get_dummies(df[xfeature],drop_first=True)
print("creditsdf:")
print(creditsdf)
# dfbank=pd.read_csv("../inputdata/bank.csv")
# dfbank.head(5)



Y =df.status
X=sm.add_constant(creditsdf)

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=.3,random_state=42)


```

```python
# Gini measurement is the probability of a random sample being classified correctly if we randomly pick a label according to the distribution in a branch. Gini impurity can be computed by summing the probability pi of an item with label i being chosen times the probability 1 ‚àí pi of a mistake in categorizing that item over all classes. It reaches its minimum (zero) when all cases in the node belong to a specific category. The Gini impurity index for a classification problem with C classes is given by

# ùê∫ùëñùëõùëñ ùêºùëõùëëùëíùë•=‚àëùëñ=1cùëùùëñ(k)(1‚àíùëùùëñ(k))
from sklearn.tree import DecisionTreeClassifier

clf_tree=DecisionTreeClassifier(criterion='gini',max_depth=4).fit(X_train,Y_train)
time_predict=clf_tree.predict(X_test)
metrics.roc_auc_score(Y_test,time_predict)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(90,120))
plot_tree(clf_tree,feature_names=X_train.columns,class_names=['Good credit','Bad credit'],filled=True)
```

```python


# Entropy is another popular measure of impurity that is used in classification trees to split a node. Assume that there are J classes labeled 1, 2, ‚Ä¶, J. The entropy at node t is given by

# ùëíùëõùë°ùëüùëúùëùùë¶=‚àí‚àëùêæùëñ=1ùëù(ùê∂ùëñ)ùëôùëúùëî2ùëù(ùê∂ùëñ)

# The value of entropy lies between 0 and 1, with a higher entropy indicating a higher impurity at the node. The following codes show how to build a classification tree using entropy splitting criteria and then displaying the tree classifier using Graphviz.

# Benefits of Decision Tree

# The benefits of building a decision tree classifier are:

# 1. Rules generated are simple and interpretable. Trees can be visualized.

# 2. Work well with both numerical and categorical data. Do not require data to be normalized or the creation of dummy variables.

# 3. Rules can help create business strategies.

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

clf_tree=DecisionTreeClassifier(criterion='entropy',max_depth=3).fit(X_train,Y_train)
time_predict=clf_tree.predict(X_test)
metrics.roc_auc_score(Y_test,time_predict)


plt.figure(figsize=(90,120))
plot_tree(clf_tree,feature_names=X_train.columns,class_names=['Good credit','Bad credit'],filled=True)
```

```python

#  we can also use a grid search mechanism to find the optimal parameters for the decision tree.
# These two parameters 1. the criteria and 2. the max depth are the parameters that can be tuned in a decision treeto find the best model.
# So, these parameters are called hyper parameters.
#use GridsearchCV for the same. 

from sklearn.model_selection import GridSearchCV
tuned_prameter=[{'criterion' :['gini','entrppy'], "max_depth":range(2,10)}]
clf=GridSearchCV(clf_tree,tuned_prameter,cv=10,scoring='roc_auc').fit(X_train,Y_train)
clf.best_score_
clf.best_params_
```
